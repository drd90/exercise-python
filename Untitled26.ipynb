```python
# Importações
import pandas as pd
import dash
from dash import dcc, html
from dash.dependencies import Input, Output
import plotly.express as px



# Inicializar a aplicação Dash
app = dash.Dash(__name__)

# Layout da aplicação
app.layout = html.Div([
    html.H1("Dashboard de Lançamentos da SpaceX", style={'textAlign': 'center'}),

    dcc.Dropdown(
        id='site-dropdown',
        options=[{'label': 'Todos os Sites', 'value': 'ALL'}] +
                [{'label': site, 'value': site} for site in df['LaunchSite'].unique()],
        value='ALL',
        placeholder='Selecione um site de lançamento',
        searchable=True
    ),

    html.Br(),

    dcc.Graph(id='success-pie-chart'),

    html.Br(),

    html.P("Intervalo de Payload (kg):"),

    dcc.RangeSlider(
        id='payload-slider',
        min=df['PayloadMass'].min(),
        max=df['PayloadMass'].max(),
        step=100,
        marks={int(x): str(int(x)) for x in range(int(df['PayloadMass'].min()), int(df['PayloadMass'].max())+1, 1000)},
        value=[df['PayloadMass'].min(), df['PayloadMass'].max()]
    ),

    html.Br(),

    dcc.Graph(id='success-payload-scatter-chart')
])

# Callback do gráfico de pizza
@app.callback(
    Output('success-pie-chart', 'figure'),
    Input('site-dropdown', 'value')
)
def update_pie_chart(selected_site):
    if selected_site == 'ALL':
        success_counts = df[df['Class'] == 1]['LaunchSite'].value_counts().reset_index()
        success_counts.columns = ['LaunchSite', 'Successes']
        fig = px.pie(success_counts, names='LaunchSite', values='Successes',
                     title='Total de Lançamentos Bem-Sucedidos por Site')
    else:
        filtered_df = df[df['LaunchSite'] == selected_site]
        outcome_counts = filtered_df['Class'].value_counts().reset_index()
        outcome_counts.columns = ['Outcome', 'Count']
        outcome_counts['Outcome'] = outcome_counts['Outcome'].map({1: 'Sucesso', 0: 'Fracasso'})
        fig = px.pie(outcome_counts, names='Outcome', values='Count',
                     title=f'Sucessos vs Fracassos no site {selected_site}')
    return fig

# Callback do gráfico de dispersão
@app.callback(
    Output('success-payload-scatter-chart', 'figure'),
    [Input('site-dropdown', 'value'),
     Input('payload-slider', 'value')]
)
def update_scatter_plot(selected_site, payload_range):
    low, high = payload_range
    filtered_df = df[(df['PayloadMass'] >= low) & (df['PayloadMass'] <= high)]

    if selected_site != 'ALL':
        filtered_df = filtered_df[filtered_df['LaunchSite'] == selected_site]

    if filtered_df.empty:
        return px.scatter(title="Nenhum dado disponível para os filtros selecionados.")

    fig = px.scatter(filtered_df,
                     x='PayloadMass',
                     y='Class',
                     color='LaunchSite',
                     title=f'Sucesso vs Payload para {selected_site}' if selected_site != 'ALL' else 'Todos os Sites')
    return fig

# Executar o servidor
if __name__ == '__main__':
    app.run(debug=True)
```
